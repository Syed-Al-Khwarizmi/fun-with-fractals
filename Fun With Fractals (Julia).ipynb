```python
# Python code for Julia Fractal 
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# setting the width, height and zoom 
# of the image to be created 
w, h, zoom = 1920,1080,1
# Because I don't feel like copying the value again and again
goldenRatio = 1.61803398875
```

```python
def generateFractal(cX=-0.8, cY=0.156, maxIter=255, moveX=0, moveY=0, colorTup=(11, 18, 8), zoom=1):
    # setting up the variables according to 
    # the equation to create the fractal 
    # The inputs to this fuction, where cX denotes the real part
    # and cY denotes the imaginary part of the equation are important.
    # Slight change(s) in the inputs can possibly generate a whole new
    # kind of a fractal, which has never been observed before.

    # creating the new image in RGB mode 
    bitmap = Image.new("RGB", (w, h), "white") 

    # Allocating the storage for the image and 
    # loading the pixel data. 
    pix = bitmap.load()

    for x in range(w): 
        for y in range(h): 
            zx = 1.5*(x - w/2)/(0.5*zoom*w) + moveX 
            zy = 1.0*(y - h/2)/(0.5*zoom*h) + moveY 
            i = maxIter 
            while zx**2 + zy**2 < 4 and i > 1: 
                tmp = zx**2 - zy**2 + cX 
                zy,zx = 2.0*zx*zy + cY, tmp 
                i -= 1

            # convert byte to RGB (3 bytes), kinda 
            # magic to get nice colors 
            pix[x,y] = (i << colorTup[0]) + (i << colorTup[1]) + i*colorTup[2]
            
    return bitmap
```

```python
# Get what the function returns
bitmap = generateFractal()
# to display the created fractal
plt.figure(figsize=(12, 18))
plt.imshow(bitmap)
```
